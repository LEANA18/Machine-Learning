```python
import pandas as pd
import numpy as np
```

```python
iris_db=pd.read_csv("Iris_data_set.csv")
data_frame=iris_db.copy()
data_frame.head()
```

```python
data_frame.describe()
```

```python
data_frame.iloc[2:3,2:4]=np.nan
data_frame.head()
```

```python
data_frame.describe()
```

```python
data_frame.dropna(inplace=True)
data_frame.head()
```

```python
#to reset index
data_frame.reset_index(drop=True,inplace=True)
data_frame.head()
```

```python
#assigning NaN value to 2nd row, 2nd and 3rd column
data_frame.iloc[2:3,2:4]=np.nan
data_frame.head()
```

```python
#to set mean value to petal_length NaN value
data_frame.petal_length.fillna(data_frame.petal_length.mean(),inplace=True)
data_frame.head()
```

```python
#to set mean value of setosa species to petal_width NaN value
#created temp data frame to store all data of setosa
temp=data_frame[data_frame.species=="setosa"]
# to take mean value of petal_width of temp data frame and assigning it to NaN of petal_width of data_frame
data_frame.petal_width.fillna(temp.petal_width.mean(),inplace=True)
data_frame.head()
```
