```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y)
plt.show()
```

```python
#to add color to the line and marker '+' to points
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y,color="red",marker="+")
plt.show()
```

```python
# to increase line width to 3
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y,color="red",marker="+",linewidth=3)
plt.show()
```

```python
# to decrease line width to 0.25
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y,color="red",marker="+",linewidth=0.25)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red")
plt.plot(x,y2,color="blue")
plt.show()
```

```python
# to show labels for that use legend()
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.legend()
plt.show()
```

```python
# to label x axis and y axis
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.show()
```

```python
# to increase the size of X and Y
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("Y",fontsize=15)
plt.xlabel("X",fontsize=15)
plt.legend()
plt.show()
```

```python
# to insert title to the graph
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("Y",fontsize=15)
plt.xlabel("X",fontsize=15)
plt.legend()
plt.title("Line Graph")
plt.show()
```

```python
# to add axis and grid
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("Y",fontsize=15)
plt.xlabel("X",fontsize=15)
plt.legend()
plt.title("Line Graph")
plt.axis([0,15,0,20000])
plt.grid()
plt.show()
```

```python
# to add text inside the graph
x=np.arange(0,10,0.25)
y=x**4
y2=x**3
plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("Y",fontsize=15)
plt.xlabel("X",fontsize=15)
plt.legend()
plt.title("Line Graph")
plt.axis([0,15,0,20000])
plt.grid()
plt.text(10,12500,"Sales",fontsize=15)
plt.show()
```
