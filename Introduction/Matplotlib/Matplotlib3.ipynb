```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
plt.pie(profit)
plt.show()
```

```python
# to add colours
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
colors=["red","magenta","pink","blue"]
plt.pie(profit,colors=colors)
plt.show()
```

```python
# to add labels
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
colors=["red","magenta","pink","blue"]
plt.pie(profit,colors=colors,labels=labels)
plt.show()
```

```python
# to show percentage of each half
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
colors=["red","magenta","pink","blue"]
plt.pie(profit,colors=colors,labels=labels,autopct="%.2f%%")
plt.show()
```

```python
#explode
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
colors=["red","magenta","pink","blue"]
explode=[0.05,0.1,0,0]
plt.pie(profit,colors=colors,labels=labels,autopct="%.2f%%",explode=explode)
plt.show()
```

```python
#clockwise direction
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
colors=["red","magenta","pink","blue"]
explode=[0.05,0.1,0,0]
plt.pie(profit,colors=colors,labels=labels,autopct="%.2f%%",explode=explode,counterclock=False,startangle=90)
plt.show()
```
