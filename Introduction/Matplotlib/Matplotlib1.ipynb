```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
# to scatter the plot
plt.scatter(x,y)
plt.show()
```

```python
# to plot the graph
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x1=[1,2,3,4,5]
x2=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[5,4,3,2,1]
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
# to draw only line ,not dotes
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
plt.plot(x1,y1)
plt.show()
```

```python
x1=[1,2,3,4,5]
x2=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[5,4,3,2,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
# to change the colour of the line
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
plt.plot(x1,y1,"r")
plt.show()
```

```python
# to change colour and pattern of line
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
plt.plot(x1,y1,"r--")
plt.show()
```

```python
# dash dot line
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
plt.plot(x1,y1,"r-.")
plt.show()
```

```python
# star shape
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
plt.plot(x1,y1,"r*")
plt.show()
```

```python
# + shape
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]
plt.plot(x1,y1,"r+")
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5,6])
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
# to make curve smooth use arange()
x=np.arange(0,10,0.1)
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```
